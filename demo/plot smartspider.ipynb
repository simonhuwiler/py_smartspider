# Plot Smart Spiders

```python
import pandas as pd
import sys
sys.path.append('../src')
import smartspider 
```

## Import DataSet

```python
#Import csv
df = pd.read_csv('demo.csv')

#Set nice name
df['name'] = df.apply(lambda row: '%s %s (%s)' % (row['firstname'], row['lastname'], row['partyAbbreviation']), axis = 1)

#show csv
df
```

## Plot

```python
smartspider.plot_dataframe(df, legend=df['name'])
```

## Plot with title

```python
smartspider.plot_dataframe(df, legend=df['name'], title='Test-Dataframe')
```

## Plot with dataframe

```python
smartspider.plot_dataframe(df, legend=df['name'], print_dataframe = True, title='Test-Dataframe')
```
